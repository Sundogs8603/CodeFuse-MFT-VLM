```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data_dict = {"model": ["LLAVA-1.5", "Qwen-VL", "Qwen-VL-14B"],
                "MMBench":[67.7, 60.6, 75.7], "MMBench-CN": [63.6, 56.7, 69.8],
                "VqaV2": [80.0, 78.2, 79.3], "GQA": [63.3, 57.5, 59.4], "TextVQA": [61.3, 63.8, 63.9],
                "VizWiz": [53.6, 38.9, 45.3]}

df = pd.DataFrame(data_dict)
```

```
labels = df.columns.delete(0).tolist()
labels.append("MMBench")


values = df.iloc[0].tolist()[1:]
values.append(df.iloc[0].tolist()[1])

# Number of variables we're plotting.
num_vars = len(labels)

# Split the circle into even parts and save the angles
# so we know where to put each axis.
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# ax = plt.subplot(polar=True)
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw the outline of our data.
ax.plot(labels, values,  label="LLAVA-1.5", color='green', linewidth=1)
# Fill it in.
ax.fill(labels, values, color='green', alpha=0.25)

values = df.iloc[1].tolist()[1:]
values.append(df.iloc[1].tolist()[1])
ax.plot(labels, values, label="Qwen-VL", color='blue', linewidth=1)
# Fill it in.
ax.fill(labels, values, color='blue', alpha=0.25)
values = df.iloc[2].tolist()[1:]
values.append(df.iloc[2].tolist()[1])
print(type(values))
ax.plot(labels, values, label="CodeFuse-VLM-14B", color='red', linewidth=1)
# Fill it in.
ax.fill(labels, values, color='red', alpha=0.25)

ax.legend(loc='lower right')
```
